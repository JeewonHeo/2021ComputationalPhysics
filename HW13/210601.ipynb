### y'' + k * exp(y) = 0
### y(0) = y(1) = 0

```python
# y = A*sin(k*x)
# y'(0) = k
# y'(0) - k = 0

def fun(x, y, p):
    k = p[0]     # Parameter setting
    f1 = y[1]    # Variable chance
    f2 = -k**2*y[0]    #Differential equation
#    return np.array([f1,f2])    # (n, m)
#    return np.vstack([f1,f2])   # (n, m)
 
    return np.vstack([y[1], -k**2 * y[0]])

def bc(ya, yb, p):    # bc(ya, yb, p), return은 n+k array가 되어야 한다.
#    ya 왼쪽
#    yb 오른쪽
#    p  parameter
    k = p[0]
    return np.array([ya[0], yb[0], ya[1] - k])

# x, y, guessing. m for x, (n, m) for y.
x = np.linspace(0, 1, 5)    # dimension 5
y = np.zeros([2,x.size])    # dimension 2, 5

y[0, 1] =  1
y[0, 3] = -1

# y[1, 1] =  1
# y[1, 3] = -1

print('y \n',y)
sol = solve_bvp(fun, bc, x, y, p=[6])
```
