```python
# 1 - ê°€)
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

def f(x,y):
    f1 = y[1]
    f2 = y[2]
    f3 = -1/x * y[2] + 1/(x**2) * y[1] + 0.1* (y[1])**3
    return np.vstack([f1,f2,f3])
  
def bc(ya, yb):    
    return np.array([ya[0], ya[2], yb[0] -1])

a = 1
b = 2
x = np.linspace(a,b,10)
y = np.zeros([3,x.size])

sol = solve_bvp(f,bc,x,y)

xa = np.linspace(a,b,100)
plt.plot(xa,sol.sol(xa)[0],'-',label='ysol')
plt.plot(xa,sol.sol(xa)[1],'--',label='ysolp')
plt.legend()
plt.grid()
plt.show()
```

```python
# 1 - ë‹¤)
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

def f(x,y):
    f1 = y[1]
    f2 = y[2]
    f3 = -2*y[2] - np.sin(y[0])
    return np.vstack([f1,f2,f3])
  
k = 0.8
def bc1(ya, yb):    
    return np.array([ya[0], yb[0], yb[1]-k])

def bc2(ya, yb):    
    return np.array([ya[0], ya[1]-k, yb[0] - 1])

a = -1
b = 0
c = 1
x1 = np.linspace(a,b,10)
x2 = np.linspace(b,c,10)
y1 = np.zeros([3,x1.size])
y2 = np.zeros([3,x2.size])
sol1 = solve_bvp(f,bc1,x1,y1)
sol2 = solve_bvp(f,bc2,x2,y2)

xa = np.linspace(a,b,100)
xb = np.linspace(b,c,100)

plt.plot(xa,sol1.sol(xa)[0],'-',label='ysol1',c='b')
plt.plot(xa,sol1.sol(xa)[1],'--',label='ysolp',c='c')
plt.plot(xb,sol2.sol(xb)[0],'-',c='b')#,label='ysol2',c='b')
plt.plot(xb,sol2.sol(xb)[1],'--',c='c')#,label='ysolp',c='c')
plt.legend()
plt.grid()
plt.show()
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

m = 20
c = 0.0005
Cd = 0.06
g = 9.80665
def f(t,y):
    x0, x1, y0, y1 = y
    v = (x1**2 + y1**2)**0.5
    f1 = x1
    f2 = -x1/m*(c*v + Cd*v**0.5)
    f3 = y1
    f4 = -y1/m*(c*v + Cd*v**0.5) - g
    return [f1,f2,f3,f4]


def bc(ya, yb):    
    return np.array([ya[0], ya[2], yb[0]-300,yb[2]-61])

a = 0
b = 9
x = np.linspace(a,b,100)
y = np.zeros([4,x.size])
sol = solve_bvp(f,bc,x,y)

print('tan(theta) =',(sol.y[2,1]-sol.y[2,0])/(sol.y[0,1]-sol.y[0,0]))
print('v0 =',(sol.y[1,0]**2 + sol.y[3,0]**2)**0.5)

plt.plot(sol.y[0],sol.y[2])
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 3
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

def f(r,y):
    f1 = y[1]
    f2 = -1/r*y[1]
    return [f1, f2]

def bc(ya, yb):    
    return np.array([ya[0],yb[0]-200])

a = 1
b = 2
x = np.linspace(a,b,100)
y = np.zeros([2,x.size])
sol = solve_bvp(f,bc,x,y)

plt.plot(sol.x,sol.y[0])
plt.grid()
plt.xlabel('r')
plt.ylabel('T(r)')
plt.show()
```
