```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def rho(x):
    if (abs(x-10) < 10): 
        return 1
    elif (abs(x+5) < 5):
        return 2
    else: 
        return 0


def f(x,y):
    return [y[1], rho(x)]

def bc(ya, yb):    
    return np.array([ya[0],yb[0]])

x = np.linspace(-30, 40, 100)
y0 = [0,0]
tspan = [-30,40]
sol = solve_ivp(f, tspan, y0, t_eval=x)
plt.plot(x, sol.y[0])
plt.show()
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
from scipy.integrate import solve_ivp


def U(r):
    return 4*(r**-12 - r**-6)

def Up(r):
    return derivative(U, r, dx=1e-6)

def f(t,y):
    f1 = y[1]
    f2 = -Up(y[0])
    return [f1, f2]

t = np.linspace(0, 3, 1000)
tspan = [0,3]
y0 = [1.05,0]
sol = solve_ivp(f,tspan,y0, t_eval=t)
plt.plot(t, sol.y[0])
plt.show()
```
