```python
# 1-1
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams.update({'font.size': 20})

data = np.genfromtxt('data/quiz5-1.txt',delimiter='    ').T
t = data[0]
y = data[1]
print('data -> \n',data)
xx = np.linspace(0, 1.5, 100)

fit1 = np.polyfit(t,y,1)  
line1 = np.poly1d(fit1)

fit2 = np.polyfit(t,y,2)  
line2 = np.poly1d(fit2)

fit4 = np.polyfit(t,y,4)  
line4 = np.poly1d(fit4)

y1 = line1(xx)
y2 = line2(xx)
y4 = line4(xx)

plt.plot(t,y,'*',markersize=20,label='Data')
plt.plot(xx,y1,label='1')
plt.plot(xx,y2,label='2')
plt.plot(xx,y4,label='4')
plt.legend()
plt.grid()
plt.show()
```

```python
# 2
from scipy.optimize import curve_fit

data = np.genfromtxt('data/quiz5-1.txt',delimiter='    ').T
t = data[0]
y = data[1]
print('data -> \n',data)

xx = np.linspace(0, 1.5, 100)
def f(t,a,b):
    return a*t**2 + b*t 

parm, cvar = curve_fit(f,t,y)

fit1 = f(xx,*parm)

plt.plot(t,y,'*',markersize=20,label='Data')
plt.plot(xx,fit1,label='fit')
plt.grid()
plt.legend()
plt.show()

g = -2*parm[0]
v0 = parm[1]
print('g =',g,'\nv0 =',v0)
```

```python
# 3

data = np.genfromtxt('data/quiz5-3.txt',delimiter=' ').T

x = data[0]
y = data[1]

def f(x,a1,b1,c1,a2,b2,c2,a3,b3,c3):
    return a1*np.exp((-(x-b1)**2)/c1**2) + a2*np.exp((-(x-b2)**2)/c2**2) + a3*np.exp((-(x-b3)**2)/c3**2)

parm, cvar = curve_fit(f,x,y,[2,3,1,8,5,1,5,7,2])
xx = np.arange(0,10,0.1)

plt.plot(xx,f(xx,*parm),linewidth=3,label='fit fnt.')
plt.plot(x,y,'.',markersize=10,label = 'Data')
plt.legend()
plt.grid()
plt.show()
print('[ a1   b1   c1   a2   b2   c2   a3   b3   c3 ]')
print(parm)
```
