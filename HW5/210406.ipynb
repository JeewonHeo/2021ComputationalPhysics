```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import *
```

```python
x = np.array([1,2,3,4,5])
y = np.array([3,6,3,6,4])

pfit = np.polyfit(x,y,1)    # n, n-1, ..., 0
line = np.poly1d(pfit)
```

```python
pfit
```

```python
xx = np.linspace(0, 6, 50)
yy = line(xx)

plt.plot(x,y,'*')    # Original data
plt.plot(xx,yy)    # Linear fit
plt.show()
```

```python
# curve_fit(f,xdata,ydata)
from scipy.optimize import curve_fit

def f(x,m,b):
    return m*x + b

parm, cvar = curve_fit(f,x,y)

yy1 = f(xx,parm[0],parm[1])
```

```python
plt.plot(x,y,'*')    # Original data
plt.plot(xx,yy)    # Linear fit
plt.plot(xx,yy1,'--')    # Second linear fit
plt.show()
```

```python
x = np.array([-2,1,4,-1,3,-4])
y = np.array([-1,2,59,4,24,-53])

pfit = np.polyfit(x,y,3)    # n, n-1, ..., 0
line = np.poly1d(pfit)
xx = np.linspace(-5, 5, 100)
yy = line(xx)

from scipy.optimize import curve_fit

def f(x,a,b,c,d):
    return a*x**3 + b*x**2 + c*x + d 

parm, cvar = curve_fit(f,x,y)

yy1 = f(xx,parm[0],parm[1],parm[2],parm[3])

plt.plot(x,y,'*')    # Original data
plt.plot(xx,yy)    # Linear fit
plt.plot(xx,yy1,'--')  
plt.show()
```

```python
# Data2.txt 를 f(x) = A*sin(B*x + C) 형태의 함수로 curve_fit 를 써서 맞추어 보시오
```

```python
import scipy as sp
data = np.genfromtxt('./data/Data2.txt',delimiter=',')    # data = loadtxt('filename',dtype=None,delimiter=',')

x = data[0]
y = data[1]

def f(x,A,B,C):
    return A*np.sin(B*x + C)

parm, cvar = curve_fit(f,x,y,[0.6,2,0])

yy1 = f(x,parm[0],parm[1],parm[2])

plt.plot(x,y,'*')    # Original data
plt.plot(x,yy1,'--') 
plt.show()
```

```python
import scipy as sp
data = np.genfromtxt('./data/oildata.dat',delimiter=',')    # data = loadtxt('filename',dtype=None,delimiter=',')
x = data[0]
y = data[1]

def veloc(t,v0,tau):
    return v0*(1 - np.exp(-t/tau))

parm, cvar = curve_fit(veloc,x,y,[5,0.1])
print(parm)
yy1 = veloc(x,parm[0],parm[1])

plt.plot(x,y,'*')    # Original data
plt.plot(x,yy1,'--') 
plt.show()
```
