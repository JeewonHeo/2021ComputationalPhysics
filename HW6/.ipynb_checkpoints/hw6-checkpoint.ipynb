```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams.update({'font.size': 20})

xdata = np.array([-1,1,3])
ydata = np.array([17,-7,-15])

x0 = xdata[0]; y0 = ydata[0]
x1 = xdata[1]; y1 = ydata[1]
x2 = xdata[2]; y2 = ydata[2]

def l0(x):
    return (x-x1)*(x-x2)/(x0-x1)/(x0-x2)

def l1(x):
    return (x-x0)*(x-x2)/(x1-x0)/(x1-x2)
    
def l2(x):
    return (x-x0)*(x-x1)/(x2-x0)/(x2-x1)
    
def p2(x):
    return y0*l0(x) + y1*l1(x) + y2*l2(x)

xx = np.arange(-3,5,0.01)
plt.plot(xx,p2(xx),label='fit fnt.')
plt.plot(xdata,ydata,'*',markersize='10',label='Data',color='black')
plt.grid()
plt.legend()
plt.show()
```

```python
def coeffts(xData,yData):
    m = len(xData)  # Number of data points
    a = yData.copy()
    for k in range(1,m):
        a[k:m] = (a[k:m] - a[k-1])/(xData[k:m] - xData[k-1])
    return a

def evalPoly(a,xData,x):
    n = len(xData) - 1  # Degree of polynomial
    p = a[n]
    for k in range(1,n+1):
        p = a[n-k] + (x -xData[n-k])*p
    return p


xx = np.arange(-3,5,0.01)
an = coeffts(xdata,ydata)
poly = evalPoly(an,xdata,xx)
plt.plot(xx,poly,label='fit fnt.')
plt.plot(xdata,ydata,'*',markersize='10',label='Data',color='black')
plt.grid()
plt.legend()
plt.show()
```

```python
def LUdecomp3(c,d,e):
    n = len(d)
    for k in range(1,n):
        lam = c[k-1]/d[k-1]
        d[k] = d[k] - lam*e[k-1]
        c[k-1] = lam
    
    return c,d,e

def LUsolve3(c,d,e,b):
    n = len(d)
    for k in range(1,n):
        b[k] = b[k] - c[k-1]*b[k-1]
    
    b[n-1] = b[n-1]/d[n-1]
    for k in range(n-2,-1,-1):
        b[k] = (b[k] - e[k]*b[k+1])/d[k]
    
    return b

def curvatures(xData,yData):
    n = len(xData) - 1
    c = np.zeros(n)
    d = np.ones(n+1)
    e = np.zeros(n)
    k = np.zeros(n+1)
    c[0:n-1] = xData[0:n-1] - xData[1:n]
    d[1:n] = 2.0*(xData[0:n-1] - xData[2:n+1])
    e[1:n] = xData[1:n] - xData[2:n+1]
    k[1:n] =6.0*(yData[0:n-1] - yData[1:n])/(xData[0:n-1] - xData[1:n]) \
            - 6.0*(yData[1:n] - yData[2:n+1])/(xData[1:n] - xData[2:n+1])
    LUdecomp3(c,d,e)
    LUsolve3(c,d,e,k)
    
    return k

def evalSpline(xData,yData,k,x):
    def findSegment(xData,x):
        iLeft = 0
        iRight = len(xData)- 1
        while 1:
            if (iRight-iLeft) <= 1: return iLeft
            i =(iLeft + iRight)//2
            if x < xData[i]: iRight = i
            else: iLeft = i
    i = findSegment(xData,x)
    h = xData[i] - xData[i+1]
    y = ((x - xData[i+1])**3/h - (x - xData[i+1])*h)*k[i]/6.0 \
            - ((x - xData[i])**3/h - (x - xData[i])*h)*k[i+1]/6.0   \
            + (yData[i]*(x - xData[i+1])                            \
            - yData[i+1]*(x - xData[i]))/h
            
    return y

xdata = np.array([0,1,2])
ydata = np.array([0,2,1])


k = curvatures(xdata,ydata)

ev=[]
xx = np.arange(-1,3,0.01)
for i in xx:
    ev.append(evalSpline(xdata,ydata,k,i))



kp = k.copy()
kp[0]=1.; kp[2]=-1.
evp=[]
for i in xx:
    evp.append(evalSpline(xdata,ydata,kp,i))

plt.plot(xx,ev,label='k0 = 0, k2 = 0')
plt.plot(xx,evp,label='k0 = 1, k2 = -1')
plt.plot(xdata,ydata,'*',markersize=10,label='Data',color='black')
plt.grid()
plt.legend()
plt.show()
```

```python
t = np.array([-250,-200,-100,0,100,300])
cp = np.array([0.0163, 0.318, 0.699, 0.870, 0.941, 1.04])

def neville(xData, yData, x):
    m = len(xData) # number of data points
    y = yData.copy()
    for k in range(1, m):
        y[0:m-k] = ((x-xData[k:m])*y[0:m-k] + 
                (xData[0:m-k] - x)*y[1:m-k+1])/(xData[0:m-k] - xData[k:m])

    return y[0]

def rational(xData, yData, x):
    m = len(xData)
    r = yData.copy()
    rOld = np.zeros(m)
    for k in range(m-1):
        for i in range(m-k-1):
            if abs(x - xData[i+k+1]) < 1.0e-9:
                return yData[i+k+1]
            else:
                c1 = r[i+1] - r[i]
                c2 = r[i+1] - rOld[i+1]
                c3 = (x - xData[i])/(x - xData[i+k+1])
                r[i] = r[i+1] + c1/(c3*(1.0 - c1/c2) - 1.0)
                rOld[i+1] = r[i+1]

    return r[0]

nev = []
rat = []
x = np.arange(-250,500)
for i in x:
    nev.append(neville(t,cp,i))
    rat.append(rational(t,cp,i))

plt.plot(x,rat,linewidth=3,label='Rational',color='red',alpha=0.8)
plt.plot(x,nev,'--',linewidth=3,label='Neville',color='green',alpha=0.7)
plt.plot(t,cp,'*',markersize=10,label='Data',color='black')
plt.grid()
plt.xlim(-250,500)
plt.legend()
plt.show()
print('''
Rational은 300이후에 위로올라가고 다항식 보간법은 아래로내려간다.
이 문제의 경우는 온도에 따른 알루미늄의 비열이므로 Rational 보간과 유사한 그래프가 맞을 것이다.
''')
```
