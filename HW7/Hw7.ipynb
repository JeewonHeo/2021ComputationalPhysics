```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams.update({'font.size': 20})
data = np.genfromtxt('../data/3.1_5.dat',delimiter=',')
xdata = data[0]
ydata = data[1]
pfit = np.polyfit(xdata,ydata,3)    # n, n-1, ..., 0
y1 = np.poly1d(pfit)
xx = np.arange(-0.5,2.5,0.01)

plt.plot(xdata,ydata,'*')
plt.plot(xx,y1(xx))
plt.grid()

root = root_scalar(y1-1,bracket=[0,2])
print('y(x) = 1 í•´ :',root.root)
```

```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt

def vb(x):
    return (np.arccos(1-x) - (1-x)*np.sin(np.arccos(1-x)))/np.pi - 0.75
xx = np.arange(1,2,0.01)

root = root_scalar(vb, method='brentq', bracket = [1.25,1.5])
print('h/r :',root.root)
plt.plot(xx,vb(xx)-0.75*vb(2))
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.optimize import root_scalar
import matplotlib.pyplot as plt

def f(x):
    return np.cos(x) - 3*np.sin(np.tan(x)-1)
    
roots = []
roots.append(root_scalar(f, method='brentq', bracket = [0.8,1]).root)
roots.append(root_scalar(f, method='brentq', bracket = [1.2,1.4]).root)
roots.append(root_scalar(f, method='brentq', bracket = [1.4,1.5]).root)

xx = np.arange(0,1.51,0.01)
plt.plot(xx,f(xx))
plt.plot(xx,np.zeros(len(xx)))
plt.grid()
plt.xlim(0,1.5)
print(roots)
```

```python
import numpy as np
from scipy.optimize import root

def c1(x,y,z):
    return (-np.tan(y)+np.tan(x))/(np.tan(z)+np.tan(y)) - 4/5

def c2(x,y,z):
    return -4*np.sin(x) - 6*np.sin(y) + 5*np.sin(z) + 3

def c3(x,y,z):
    return 4*np.cos(x) + 6*np.cos(y) + 5*np.cos(z) - 12

def func(x):
    return [c1(x[0],x[1],x[2]),c2(x[0],x[1],x[2]),c3(x[0],x[1],x[2])]

#x0 = [4/5,-3,12]
x0 = [1,1,1]
sol = root(func,x0)
ans = sol.x
print(f"theta1 = {ans[0]}\ntheta2 = {ans[1]}\ntheta3 = {ans[2]}")
print(f"check ans -> \n(c1(ans) c2(ans) c3(ans)) = {c1(*ans).round(),c2(*ans).round(),c3(*ans).round()}")
```
